Calculator

```python
print("***** Calculator *****")
print("1) Sum")
print("2) Sub")
print("3) Mul")
print("4) Div")


calculator = int(input("Select Option: "))

x = int(input("Enter the value of a: "))
y = int(input("Enter the value of b: "))

match calculator:
  case 1:
    print("Sum is: ")
    Sum = x+y
    print(Sum)
  case 2:
    print("Sub is: ")
    Sub = x-y
    print(Sub)
  case 3:
    print("Mul is: ")
    Mul = x*y
    print(Mul)
  case 4:
    print("Div is: ")
    Div = x/y
    print(Div)

```
